```python
import random   #importing random module

target=random.randint(1,100) #randomly selecting the number to guess

#Creating a function to compare the user guess with selected number
def compare():
  choice=int(input('Please guess a number between 1 to 100 - '))
  if choice==target:
    return 0
  elif choice<target:
    return 'Too low.'
  else:
    return 'Too high.'

#Letting the user to select the difficulty mode - Easy mode gets 10 lives and Hard mode gets 5 lives
def mode():
  mod=input("Please enter the mode to play 'Easy(E)' or 'Hard(H)' - ").lower()
  if mod=='e':
    return 10
  elif mod=='h':
    return 5
  else:
    print('Invalid choice')
    mode()

#Creating the main function to start the game and calling other functions.
def play():
  chances=mode()
  for _ in range(0,chances):
    result=compare()
    if result==0:
      print('Congratulations, you have guessed the right number.')
      break
    print(result,'\n','You have',chances-_-1,'chances to guess the answer.')
    if chances-_-1==0:
      print('Games Over.')

while input('Do you want to play the game? Y/N - ').lower()=='y':
  play()
```
