```python
#Creating a function to encrypt the Word 
def encrypt(i,shift,alphabet):
    if i.isalpha():
        if i.isupper():
           new.append(alphabet[alphabet.index(i.lower())+shift].upper())
        elif i.islower():
            new.append(alphabet[alphabet.index(i)+shift])
    else:
        new.append(i)
    return new

#Creating a function to decrypt the Word     
def decrypt(i,shift,alphabet):
    if i.isalpha():
        if i.isupper():
            new.append(alphabet[alphabet.index(i.lower(),26)-shift].upper())
        elif i.islower():
            new.append(alphabet[alphabet.index(i,26)-shift])
    else:
        new.append(i)
    return new
        
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z','a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
yes=True
while yes:
    choice = int(input('Enter 1 for encoding and 2 for decoding \n'))
    word=input('Please enter the word to encode/decode - ')
    shift=int(input('Shift number - '))
    shift%=26
    new=[]
    if choice==1:
        for i in word:
            f=encrypt(i,shift,alphabet)  #Calling the encrypting function
    else:
        for i in word:
            f=decrypt(i,shift,alphabet)  #Calling the decrypting function
    final="".join(f)
    print(final)
    y=input('Do you want to do another operation - Y/N? ').upper()
    if y=='N':
        yes=False
        print('Goodbye')
```
